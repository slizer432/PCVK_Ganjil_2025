```python
R, G, B = img[:,:,0], img[:,:,1], img[:,:,2]

# Gabungkan Red-Blue (hilangkan Green)
RB = img.copy()
RB[:,:,1] = 0   # channel G di-nol-kan

# Gabungkan Green-Blue (hilangkan Red)
GB = img.copy()
GB[:,:,0] = 0   # channel R di-nol-kan

plt.subplot(1,2,1); plt.imshow(RB); plt.title("Red + Blue")
plt.subplot(1,2,2); plt.imshow(GB); plt.title("Green + Blue")
plt.show()
```

```python
canvas = np.zeros((300,300,3), dtype=np.uint8)

for _ in range(5):
    x1, y1 = np.random.randint(0, 200, size=2)
    x2, y2 = x1 + np.random.randint(20,100), y1 + np.random.randint(20,100)
    color = tuple(np.random.randint(0,256, size=3).tolist())
    cv.rectangle(canvas, (x1,y1), (x2,y2), color, -1)

plt.imshow(canvas[:,:,::-1])
plt.title("5 Kotak Acak")
plt.show()
```
